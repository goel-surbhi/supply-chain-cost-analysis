### Expected Billing rate

```python
df1['fwd_fixed_zone_key'] = 'fwd_' + df1['Zone as per X'] + '_fixed'
```

```python
df1 = pd.merge(df1, c_rates_new, left_on = 'fwd_fixed_zone_key', right_on = 'key', how = 'left')
```

```python
df1.rename(columns = {"Rate":'Fixed zone rate'}, inplace = True)
df1.drop('key', axis = 1, inplace = True)
```

```python
df1
```

```python
df1['fwd_add_zone_key'] = 'fwd_' + df1['Zone as per X'] + '_additional'
```

```python
df1 = pd.merge(df1, c_rates_new, left_on = 'fwd_add_zone_key', right_on = 'key', how = 'left')
```

```python
df1.rename(columns = {"Rate":'Additional zone rate'}, inplace = True)
df1.drop('key', axis = 1, inplace = True)
```

```python
df1['rto_fixed_zone_key'] = 'rto_' + df1['Zone as per X'] + '_fixed'
```

```python
df1 = pd.merge(df1, c_rates_new, left_on = 'rto_fixed_zone_key', right_on = 'key', how = 'left')
```

```python
df1.rename(columns = {"Rate":'Fixed Rto zone rate'}, inplace = True)
df1.drop('key', axis = 1, inplace = True)
```

```python
df1['rto_add_zone_key'] = 'rto_' + df1['Zone as per X'] + '_additional'
```

```python
df1 = pd.merge(df1, c_rates_new, left_on = 'rto_add_zone_key', right_on = 'key', how = 'left')
```

```python
df1.rename(columns = {"Rate":'Additional Rto zone rate'}, inplace = True)
df1.drop('key', axis = 1, inplace = True)
```

```python
df1
```

```python
df1['Type of Shipment'].unique()
```

```python
df1['Fixed Rto zone rate'] = np.where(df1['Type of Shipment'] == 'Forward charges', 0, df1['Fixed Rto zone rate'])
df1['Additional Rto zone rate'] = np.where(df1['Type of Shipment'] == 'Forward charges', 0, df1['Additional Rto zone rate'])
```

```python
df1
```

```python
# expected price = fixed_zone_rate + additional_zone_rate*(weight_slab - 0.5)*2 + Fixed Rto zone rate + Additional Rto zone rate*(weight_slab - 0.5)*2
```

```python
df1['Expected Bill X'] = df1['Fixed zone rate'] + (df1['Additional zone rate']*(df1['Weight Slab X(KG)']-0.5)*2) + df1['Fixed Rto zone rate'] + (df1['Additional Rto zone rate']*(df1['Weight Slab X(KG)']-0.5)*2)
```

```python
df1.columns
```

```python
final_df = df1.drop(['fwd_fixed_zone_key',
       'Fixed zone rate', 'fwd_add_zone_key', 'Additional zone rate',
       'rto_fixed_zone_key', 'Fixed Rto zone rate', 'rto_add_zone_key',
       'Additional Rto zone rate'], axis = 1)
```

```python
final_df
```

```python
final_df['Bill Difference'] = final_df['Expected Bill X'].round(2) - final_df['Billing Amount (Rs.)'].round(2)
```

```python
# Output Data 1 Complete
# final_df.to_csv
```

## Output Data 2

```python
final_df['Result'] = np.where(final_df['Bill Difference'] == 0, 'Correctly Charged',
                             np.where(final_df['Bill Difference']>0, 'Under Charged', 
                                     np.where(final_df['Bill Difference']<0, 'Over Charged', None)))
```

```python
final_df['Result'].unique()
```

```python
final_summary = final_df.groupby('Result').agg({'Order ID':'count', 'Billing Amount (Rs.)':'sum'})
```
